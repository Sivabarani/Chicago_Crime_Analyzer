Fill Missing values

```python
# 'Location Description' has missing (empty) values. 
# As the exact location for these entries is unavailable, 
# the missing values have been filled with 'UNKNOWN' to ensure consistency in the dataset.
df['Location Description'] = df['Location Description'].fillna('UNKNOWN')
```

```python
# Group by relevant features and fill missing 'Ward' with the mode of each group
grouped_mode = df.groupby(['District', 'Primary Type'])['Ward'].apply(lambda x: x.mode()[0] if not x.mode().empty else pd.NA)

# Use the mode to fill missing values
df['Ward'] = df.apply(
    lambda row: grouped_mode.loc[(row['District'], row['Primary Type'])] 
    if pd.isna(row['Ward']) else row['Ward'], axis=1
)

# Group by relevant features and fill missing 'Ward' with the mode of each group
grouped_mode = df.groupby(['District'])['Ward'].apply(lambda x: x.mode()[0] if not x.mode().empty else pd.NA)

# Use the mode to fill missing values
df['Ward'] = df.apply(
    lambda row: grouped_mode.loc[(row['District'])] 
    if pd.isna(row['Ward']) else row['Ward'], axis=1
)
```

```python
# Group by relevant features and fill missing 'ommunity' with the mode of each group
grouped_mode = df.groupby(['District', 'Primary Type'])['Community Area'].apply(lambda x: x.mode()[0] if not x.mode().empty else pd.NA)

# Use the mode to fill missing values
df['Community Area'] = df.apply(
    lambda row: grouped_mode.loc[(row['District'], row['Primary Type'])] 
    if pd.isna(row['Community Area']) else row['Community Area'], axis=1
)

# Group by relevant features and fill missing 'Community' with the mode of each group
grouped_mode = df.groupby(['District'])['Community Area'].apply(lambda x: x.mode()[0] if not x.mode().empty else pd.NA)

# Use the mode to fill missing values
df['Community Area'] = df.apply(
    lambda row: grouped_mode.loc[(row['District'])] 
    if pd.isna(row['Community Area']) else row['Community Area'], axis=1
)
```
